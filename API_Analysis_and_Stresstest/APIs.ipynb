```python
import requests
```

# Aoe2

## Strings

```python
# strings
r = requests.get("https://aoe2.net/api/strings?game=aoe2de&language=en")
r = r.json()
r
```

## Leaderboard

```python
# leaderboard
r = requests.get("https://aoe2.net/api/leaderboard?game=aoe2de&leaderboard_id=3&start=1&count=10")
r = r.json()
r
```

## Lobbies

```python
# lobbies
r = requests.get("https://aoe2.net/api/lobbies?game=aoe2de")
r = r.json()
r
```

## Player Last Match

```python
# player last match
r = requests.get("https://aoe2.net/api/player/lastmatch?game=aoe2de&steam_id=76561198313563347")
r = r.json()
r
```

## Player Match History

```python
# player match history
r = requests.get("https://aoe2.net/api/player/matches?game=aoe2de&steam_id=76561198313563347&start=0&count=2")
r = r.json()
r
```

## Player Rating History

```python
# player rating history
r = requests.get("https://aoe2.net/api/player/ratinghistory?game=aoe2de&leaderboard_id=3&steam_id=76561198313563347&count=5")
r = r.json()
r
```

## Match History 

```python
# match history
# count=1: <1000
r = requests.get("https://aoe2.net/api/matches?game=aoe2de&count=1&since=1596775000")
r = r.json()
r
```

## Match Details

```python
# match details
r = requests.get(" https://aoe2.net/api/match?uuid=1589fde4-faa9-8645-844e-422e11599cbe")
r = r.json()
r
```
